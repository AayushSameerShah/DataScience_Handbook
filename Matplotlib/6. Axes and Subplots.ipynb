### Easy... Now, as this gives us more control over the placement of the Plots, we don't often have to use this as it makes us to write more code.

We can use *Subplots* which will let us t create the plots by its own with less line of code

```python
fig = plt.figure()

ax1 = plt.subplot(2,2,1)
ax2 = plt.subplot(2,2,2)
ax3 = plt.subplot(2,2,3)
ax4 = plt.subplot(2,2,4)
```

#### Also we can put the TEXT!

```python
fig = plt.figure()

ax1 = plt.subplot(2,2,1)
ax2 = plt.subplot(2,2,2)
ax3 = plt.subplot(2,2,3)
ax4 = plt.subplot(2,2,4)


ax1.text(.5, .5, '1st Plot')
ax2.text(.5, .5, '1st Plot')
ax3.text(.5, .5, '1st Plot')
ax4.text(.5, .5, '1st Plot')
```

#### Control space between the plots

```python
fig = plt.figure()

ax1 = plt.subplot(2,2,1)
ax2 = plt.subplot(2,2,2)
ax3 = plt.subplot(2,2,3)
ax4 = plt.subplot(2,2,4)


ax1.text(.5, .5, '1st Plot')
ax2.text(.5, .5, '1st Plot')
ax3.text(.5, .5, '1st Plot')
ax4.text(.5, .5, '1st Plot')

fig.subplots_adjust(hspace= 0.4, wspace= 0.4)
```

### GridSpec
The most aggressive and amazing feature for plotting


```python
grid = plt.GridSpec(4,4)
```

```python
left = plt.subplot(grid[:3, 0])
box = plt.subplot(grid[:3,1:])
bottom = plt.subplot(grid[3, 1:])
```

Looks so nice!!!!
Now, let's sharpen it up...

```python
fig = plt.figure(figsize= (5,5))
grid = plt.GridSpec(4,4)

left = plt.subplot(grid[:3, 0])
box = plt.subplot(grid[:3,1:])
bottom = plt.subplot(grid[3, 1:])
```

Removing tick labels...

```python
fig = plt.figure(figsize= (5,5))
grid = plt.GridSpec(4,4)

left = plt.subplot(grid[:3, 0])
box = plt.subplot(grid[:3,1:])
bottom = plt.subplot(grid[3, 1:])

box.set(xticklabels= [], yticklabels= []);
left.set(xticklabels= []);
bottom.set(yticklabels= []);
```

```python
fig = plt.figure(figsize= (5,5))
grid = plt.GridSpec(4,4)

left = plt.subplot(grid[:3, 0])
box = plt.subplot(grid[:3,1:])
bottom = plt.subplot(grid[3, 1:])

box.set(xticklabels= [], yticklabels= []);
left.set(xticklabels= []);
bottom.set(yticklabels= []);


x = np.random.randint(0,100,500)
y = np.random.randint(0,100,500)
box.plot(x, y,'ok', markersize= 1.5, alpha= 0.5)
left.hist(y, orientation= 'vertical');
left.invert_yaxis();
bottom.hist(x);
```
