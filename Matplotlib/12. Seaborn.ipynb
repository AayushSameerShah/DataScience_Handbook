### Seaborn plots... more...

```python
import pandas as pd
```

```python
d1 = pd.DataFrame(np.random.normal(size= (2000,2)), columns= ['x','y'])
```

```python
plt.hist(d1.x, alpha= 1);
plt.hist(d1.y, alpha= 0.6);
plt.legend(['x','y'])
```

### KDE!

```python
sns.kdeplot(d1.x, shade= True);
sns.kdeplot(d1.y, shade= True);
plt.legend(['a','b'])
```

### hist + kde = distplot

```python
sns.histplot(d1.x, kde= True);
sns.histplot(d1.y, kde= True);
```

### If we would have passed the whole DataFrame in `kdeplot` it would have returned the 2D graph ( does not work in all plots)

```python
sns.kdeplot(x= 'x', y= 'y', data= d1)
```

### jointplot


```python
sns.jointplot(x= "x", y= "y", data= d1, kind= 'kde')
```

```python
with sns.axes_style('ticks'):
    sns.jointplot(x= "x", y= "y", data= d1, kind= 'kde')
```
