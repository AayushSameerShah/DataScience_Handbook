```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
space = np.linspace(1,25,100)
```

#### Now, I want to plot two figures like two plots together

## Common approach

```python
fig = plt.figure()
plt.plot(space, np.cos(space)) # Data 1
plt.plot(space, np.sin(space)) # Data 2
plt.plot(space, np.tan(space)) # Data 3
```

### No! I want to see them both in the different plot!! This will look messy!

<font size= 5> Yea, you can create `different figures` for different plots and then show them one by one... but that is not the case! That will require more keystrokes and will not be in the same plot!!

> What we want:
# We need to get the different plots in the same plot but not in the single plot!

### Perfect approach

```python
fig = plt.figure()

plt.subplot(3,1,1)
plt.plot(space, np.cos(space)) # Data 1

plt.subplot(3,1,2)
plt.plot(space, np.sin(space)) # Data 2

plt.subplot(3,1,3)
plt.plot(space, np.tan(space)) # Data 3
```

<font face='consolas'> <b> Now let's talk about the problem. </b> </font>

    In this kind of style where we define the fig and the plot the graphs in side it... is 
    okay for the limited scope.
    
    fig, holds the plots and plots them. But the problem arises when we want multiple figs 
    and then need to change the previous fig. Now what?
    
    See...
    
    fig1 = ...
    plot...
    plot...
    
    fig2 = ...
    plot...
    plot...
    
    Now I want to change in the fig1... now what?
    
## Here comes the `Object Oriented Approach`
    

```python
fig, ax = plt.subplots(2)

ax[0].plot(space, np.sin(space))
ax[1].plot(space, np.cos(space))
```

    Also notice the 's' in subplots.

### Just for the scatter plots, there is no `a new` plot type, but just the same `.plot()` with the third arg (fmt) `'o'`
And also notice that it is not `linestlye` parameter

* Do note that, there is one "Original" way to create the scatter plots, with `.scatter()` (Will discuss in the moment)

```python
spacex = space
spacey = np.sin(space)
```

```python
plt.plot(spacex, spacey, 'o')
```

```python
plt.figure(figsize= (15,5))
plt.plot(spacex, spacey, '.-.')
```

```python
plt.figure(figsize= (15,5))
plt.plot(spacex, spacey, 'o-.', markerfacecolor= 'red')
```

# The `original` way of creating the `scatter plots` is... `.scatter()`

```python
plt.scatter(spacex, spacey)
```

#### You can't make the `.scatter()` a fool !

```python
plt.scatter(spacex, spacey, marker= '--')  # Tried to make the scatter to work as normal line !!! But it is not fool!
```

## So what's the difference?

    Basically scatter is more powerful in terms of handling the "Individual" points of the data.
    So, you can access each data in the plot inivudually which was not possible in the plot - scatter plot!

```python
temp1 = np.random.randint(1,100,100)
temp2 = np.random.randint(1,100,100)
```

```python
plt.figure(figsize= (10,10))
plt.scatter(temp1, temp2)
```

### Actully control the individual points in action

```python
temp1 = np.random.randint(1,100,100)
temp2 = np.random.randint(1,100,100)
color = np.random.randint(1,100,100)
```

```python
plt.figure(figsize= (10,10))
plt.scatter(temp1, temp2, c= color)
```

```python
# Now putting the size
size= np.random.randint(0,100,100)

plt.figure(figsize= (10,10))
plt.scatter(temp1, temp2, c= color, sizes= size)
```

```python
# Now putting the size
size= np.random.randint(0,100,100)

plt.figure(figsize= (10,10))
plt.scatter(temp1, temp2, c= color, sizes= size)

#Now placing the colorbar
plt.colorbar()
```

    The errorbar basicaly plots the data in the .plot() format, the main thing is, it is able to plot the vertical line

```python
spacex = np.linspace(1,100,10)
spacey = np.sin(spacex)
```

See...?

### Now, let's try some more concepts on `.fill_between()` - which is quite amazing.
Again, for notes - see the book.

```python
fig, ax = plt.subplots(2,2, figsize= (20,10))

ax[0][0].plot(spacex, spacey)
ax[0][0].fill_between(spacex, spacey)
ax[0][0].set_title("ax[0][0]")

ax[0][1].plot(spacex, spacey)
ax[0][1].fill_between(spacex, 1, spacey)
ax[0][1].set_title("ax[0][1]")

ax[1][0].plot(spacex, spacey)
ax[1][0].fill_between(spacex, 0, spacey)
ax[1][0].set_title("ax[1][0]")

ax[1][1].plot(spacex, spacey)
ax[1][1].fill_between(spacex, -1, spacey)
ax[1][1].set_title("ax[1][1]")
```

### There... again...
The broadcasting rule applied... to give the different numbers of the hills and valleys you can give the individual values to them!

```python
plt.plot(spacex, spacey)
plt.fill_between(spacex, [1,-2,3,-5,2,-7,6,-8,1,-3], spacey)
```
