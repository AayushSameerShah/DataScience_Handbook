### Simple...

#### Now, customizing...

```python
space = np.linspace(0, 10, 100)
I = np.sin(np.cos(space) * space[:, np.newaxis])

plt.imshow(I)
plt.colorbar(extend = "both")
plt.clim(-1, 1)
```

### Discrete contours

```python
plt.imshow(I, cmap= plt.cm.get_cmap('Blues', 4))
plt.colorbar()
plt.clim(-1, 1)
```

```python
plt.imshow(I, cmap= plt.cm.get_cmap('Blues', 4))
plt.colorbar(ticks= [-1, -.5, 0, .5, 1])
plt.clim(-1, 1)
```

```python
from sklearn.manifold import Isomap
```

```python
iso = Isomap(n_components= 2)
```

```python
projection = iso.fit_transform(DIGITS0_to_5.data)
```

```python
projection.shape
```

```python
projectionX, projectionY = projection.T
```

```python
plt.scatter(projectionX, projectionY, lw= 0.01, c= DIGITS0_to_5.target, cmap= 'jet')
plt.colorbar()
```

```python
plt.scatter(projectionX, projectionY, lw= 0.01, c= DIGITS0_to_5.target, cmap= plt.cm.get_cmap('jet', 6))
plt.colorbar()
```
