###  Now, let's plot the real data!

````python
spacex = np.linspace(0, 10, 1000)
spacey = np.sin(spacex)
spacez = np.cos(spacex)
````

#### Normal 2D

````python
ax = plt.axes()
ax.plot(spacex, spacey)
````

#### Just amazing!

````python
ax = plt.axes(projection= '3d')
ax.plot3D(spacez, spacey, spacex)
````

````python
randx = np.random.randn(100)
randy = np.sin(randx) + 0.1 * np.random.randn(100)
randz = np.cos(randx) + 0.1 * np.random.randn(100)
````

````python
ax = plt.axes(projection= '3d')
ax.plot3D(spacez, spacey, spacex)
ax.scatter3D(randz, randy, randx, c=randz, cmap= 'jet')
````

#### 3D Contours?

````python
x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)
````

````python
X, Y = np.meshgrid(x,y)
````

````python
plt.figure(figsize= (10,10))
plt.plot(X, Y, 's', markersize=15);
````

### So that is the mashgrid...
Now, on top of that, we will build the Z axe

````python
from mpl_toolkits import mplot3d
%matplotlib notebook
````

Creating the simple

````python
Z = np.sin( X )
````

````python
fig = plt.figure(figsize= (10,10))
ax = plt.axes(projection = '3d')

ax.contour3D(X, Y, Z, 100)
````

Complex

````python
Z = np.sin( np.sqrt(X ** 2 + Y ** 2) )
````

````python
fig = plt.figure(figsize= (10,10))
ax = plt.axes(projection = '3d')

ax.contour3D(X, Y, Z, 100)
````

````python
fig = plt.figure(figsize= (5,5))
ax = plt.axes(projection = '3d')

ax.contour3D(X, Y, Z, 100)
ax.set_xlabel('x')              # See the Lables!
ax.set_ylabel('y')
ax.set_zlabel('z');
````

# How amazing!!
Now, we will look forward to the wireframes and the surface plots

````python
fig = plt.figure()
ax = plt.axes(projection= '3d')

ax.plot_wireframe(X,Y,Z)
````

````python
fig = plt.figure()
ax = plt.axes(projection= '3d')

ax.plot_surface(X,Y,Z, cmap= 'seismic')
````

````python
fig = plt.figure()
ax = plt.axes(projection= '3d')

ax.scatter3D(x, y, z)
````
