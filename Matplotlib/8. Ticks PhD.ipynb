### Controling the amount of ticks

```python
fig, axes = plt.subplots(4,4, sharex= True, sharey= True, figsize= (10,10))
```

Look at that bullshit... 

```python
fig, axes = plt.subplots(4,4, sharex= True, sharey= True, figsize= (10,10))

for axe in axes.flat:
    axe.xaxis.set_major_locator( plt.MaxNLocator(2) )
    axe.yaxis.set_major_locator( plt.MaxNLocator(2) )
```

```python
fig, axes = plt.subplots(4,4, sharex= True, sharey= True, figsize= (10,10))

for axe in axes.flat:
    axe.xaxis.set_major_locator( plt.MaxNLocator(2) )
    axe.yaxis.set_minor_locator( plt.MaxNLocator(2) )
```

# A control over what should be displayed

```python
base = np.linspace(0, 3 * np.pi, 1000)

spacex = np.sin(base)
spacey = np.cos(base)
```

```python
fig, ax = plt.subplots()
ax.plot(base, spacex)
ax.plot(base, spacey)

ax.axis('equal')
ax.grid(False)
```

```python
fig, ax = plt.subplots()
ax.plot(base, spacex)
ax.plot(base, spacey)

ax.axis('equal')
ax.grid(False)

ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
```

    No, we want to show the Custom... we will use the funcFormatter !

```python
def printWhen(value, _):
    value = int(round(2 * value / np.pi))
    
    if value == 0:
        return 0
    elif value == 1:
        return r"$\pi/2$"
    elif value == 2:
        return r"$\pi$"
    else:
        return "CHACHA"
```

```python
fig, ax = plt.subplots()
ax.plot(base, spacex)
ax.plot(base, spacey)

ax.axis('equal')
ax.grid(False)

ax.xaxis.set_major_formatter(plt.FuncFormatter(printWhen))
```
