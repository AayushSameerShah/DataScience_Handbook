### 2D histograms ( call it heatmaps? )

```python
x = np.random.normal(0,100,10000)
y = np.random.normal(0,100,10000)

plt.hist2d(x, y, bins= 100, cmap= 'Blues');
plt.colorbar()
```

### Likewise `Hexbin`

```python
plt.hexbin(x,y, cmap= 'Blues', gridsize= 20)
plt.colorbar()
```

### KDE

```python
xy = np.vstack([x,y])
```

```python
from scipy.stats import gaussian_kde
```

```python
kde = gaussian_kde(xy)
```

#### KDE obj is ready for the `xy` dataset.
Now, it is like a `groupby` and ready to apply the function for the further exploration.
_____
Let's create the data from that KDE

```python
spacex = np.linspace(-3.5, 3.5, 40)
spacey = np.linspace(-6, 6, 40)
```

```python
X, Y = np.meshgrid(spacex, spacey)
```

    Here X.ravel() is used -- to flatten the array

```python
Z = kde.evaluate(np.vstack([X.ravel(), Y.ravel()]))
```

```python
plt.imshow(Z.reshape(X.shape),
          origin= 'lower',
          cmap= 'Blues', aspect= 'auto',
          extent= [-3.5,3.5,-6,6])
```
