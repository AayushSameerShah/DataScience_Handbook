### So to do that, we have to *(obviously)* prepare 3 axes - `x, y, z`
We will accomplish that with `np.meshgrid()`

It actully help to create a mash with which we can access the coordinates in the 2D array... so... just see the example below.

#### `np.mashgrid()` will help us to create the axis ( grid combination )

```python
_1, _2 = np.meshgrid([0,1,2], [0,1,2])
```

```python
_1
```

```python
_2
```

```python
for l, r in zip(_1, _2):
    print(r[0], l[0], sep='')
    print(r[1], l[1], sep='')
    print(r[2], l[2], sep='')
```

# Actual Example from the book

```python
spacex = np.linspace(0,5,50)
spacey = np.linspace(0,5,40)
```

```python
X, Y = np.meshgrid(spacex, spacey)
```

```python
Z = np.sin(X) ** 10 + np.cos(10 + Y * X) + np.cos(X)
```

```python
plt.contour(X, Y, Z, 20, cmap= 'seismic')
```

###  `.contourf()`

```python
plt.contourf(X, Y, Z, 20, cmap= 'seismic')
plt.colorbar()
```

### To avoid the computation...
Use the simple image!

```python
plt.imshow(Z)
```

#### Really!
But it looks upside down... 
<br> Oh just use the `origin= 'lower'`

```python
plt.imshow(Z, origin= 'lower')
```

#### Still not satisfied? 
Use the same colormap

```python
plt.imshow(Z, origin= 'lower', cmap= 'seismic')
```

#### Compare both ways...

```python
fig, axes = plt.subplots(2, figsize= (5,10))

axes[0].contourf(X, Y, Z, 20, cmap= 'seismic')
axes[1].imshow(Z, origin= 'lower', cmap= 'seismic')
```

#### How, cool! Just with one axes `z` we have plotted the plot successfully.

### Contours with Lables

```python
spacex = np.linspace(0,5,50)
spacey = np.linspace(0,5,40)

X, Y = np.meshgrid(spacex, spacey)

Z = np.sin(X) ** 10 + np.cos(10 + Y * X) + np.cos(X)
```

```python
plt.contour(X, Y, Z, 3)
```

```python
contours = plt.contour(X, Y, Z, 3, colors= 'black')
```

```python
contours
```

```python
plt.clabel(contours, inline= True, fontsize= 25)
plt.imshow(Z, extent= [5,0,5,0], origin= 'lower',
                       alpha= 0.5)
plt.colorbar()
```
