### Some real world

```python
plt.figure(figsize= (8,8))
M = Basemap(projection= 'lcc', lat_0= 20, lon_0= 78, width= 8E6, height= 8E6)
M.etopo(scale= 0.5) # Instead of 'Bluemarble'
```

```python
plt.figure(figsize= (8,8))
M = Basemap(projection= 'lcc', lat_0= 20, lon_0= 78, width= 8E6, height= 8E6)
M.etopo(scale= 0.5) # Instead of 'Bluemarble'

x, y = M(70, 23)
plt.plot(x, y, marker= 'o', color= 'pink')
plt.text(x, y, " Gujarat", color= 'k', fontsize= 15);
```
