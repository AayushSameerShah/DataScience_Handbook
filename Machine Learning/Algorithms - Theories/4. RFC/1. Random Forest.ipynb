##### `Manual` Bagging and RFC

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
```

```python
tree = DecisionTreeClassifier()
bag = BaggingClassifier(tree)
```

    base_estimator : object, default=None
        The base estimator to fit on random subsets of the dataset.
        If None, then the base estimator is a decision tree.

```python
accuracy_score(y, BaggingClassifier().fit(X, y).predict(X))
```

### ↑ See? ↓
You can give Tree, or other - otherwise will take `Tree` as default.

```python
accuracy_score(y, BaggingClassifier(tree).fit(X, y).predict(X))
```

### 

### 

`n_estimators` is about number of models to take as the candidates for voting.

```python
tree = DecisionTreeClassifier()
bag = BaggingClassifier(tree, n_estimators= 200)

bag.fit(Xtrain, ytrain)
accuracy_score(ytest, bag.predict(Xtest))
```

### Now, let's see the distribution - how does it look after all 200 estimators... 

```python
bag = BaggingClassifier(n_estimators= 200).fit(X, y)
plt.scatter(X[:, 0], X[:, 1], c = y, cmap= 'rainbow')
plt.axis('off');

xlim = plt.xlim()
ylim = plt.ylim()

xx, yy = np.meshgrid(np.linspace(*xlim, num= 200),
                     np.linspace(*ylim, num= 200))
xy = np.vstack([xx.ravel(), yy.ravel()]).T
pred = bag.predict(xy).reshape(xx.shape)

total_clasees = len(np.unique(y))
plt.contourf(xx, yy, pred, levels= np.arange(total_clasees + 1) - 0.5, alpha= 0.3, cmap= 'seismic')
```

###### `Automatic` - RFC 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples= 200, n_features= 2, centers= 4, cluster_std= 2)
plt.scatter(X[:,0], X[:,1], c=y, cmap= 'rainbow')
```

```python
rfc = RandomForestClassifier(n_estimators= 200).fit(X, y)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap= 'rainbow')

xx = np.linspace(*plt.xlim(), num= 200)
yy = np.linspace(*plt.ylim(), num= 200)
xx, yy = np.meshgrid(xx, yy)
xy = np.c_[xx.ravel(), yy.ravel()]

pred = rfc.predict(xy).reshape(xx.shape)
plt.contourf(xx, yy, pred, alpha= 0.3, levels= np.arange(len(np.unique(y)) + 1) - 0.5, cmap= 'seismic')
```

```python
accuracy_score(y, rfc.predict(X))
```

#### With Training splits... 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
rfc = RandomForestClassifier(n_estimators= 200).fit(X_train, y_train)
```

```python
accuracy_score(y_test, rfc.predict(X_test))
```
