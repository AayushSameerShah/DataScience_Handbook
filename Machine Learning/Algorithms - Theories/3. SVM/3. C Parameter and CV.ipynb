# Here... (again)

The data will be in the form of mixed, so it will be hard for the SVM to create a sepearate line... but let's experiment a little bit.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set();

from sklearn.datasets import make_blobs
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVC
```

```python
plt.style.use('dark_background')
```

```python
X, y = make_blobs(n_samples= 500, centers= 2, cluster_std= 3)
plt.axes().grid(False)
plt.scatter(X[:, 0], X[:, 1], c= y, cmap= 'rainbow')
```

```python
fig, axes = plt.subplots(2, 2, figsize= (20, 20))

for ax, C in zip(axes.ravel(), [1, 100, 1000, 10000]):
    model = SVC(kernel= 'linear', C= C)
    model.fit(X, y)
    
    ax.grid(False)
    ax.scatter(X[:, 0], X[:, 1], c= y, cmap= 'rainbow')
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    xx = np.linspace(xlim[0], xlim[1], 500)
    yy = np.linspace(ylim[0], ylim[1], 500)

    xx, yy = np.meshgrid(xx, yy)
    xy = np.vstack([xx.ravel(), yy.ravel()]).T
    line = model.decision_function(xy).reshape(xx.shape)

    ax.contour(xx, yy, line, levels= [-1, 0, 1], linestyles= ['--', '-', '--'], colors= 'w')
    ax.set_title(f'C = {C}');
```

# Same with 1, 100, 1000, 10000 ?
Now, let's try with `less strict` values...

```python
fig, axes = plt.subplots(2, 2, figsize= (20, 20))

for ax, C in zip(axes.ravel(), [0.1, 0.01, 0.001, 0.0001]):
    model = SVC(kernel= 'linear', C= C)
    model.fit(X, y)
    
    ax.grid(False)
    ax.scatter(X[:, 0], X[:, 1], c= y, cmap= 'rainbow')
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    xx = np.linspace(xlim[0], xlim[1], 500)
    yy = np.linspace(ylim[0], ylim[1], 500)

    xx, yy = np.meshgrid(xx, yy)
    xy = np.vstack([xx.ravel(), yy.ravel()]).T
    line = model.decision_function(xy).reshape(xx.shape)

    ax.contour(xx, yy, line, levels= [-1, 0, 1], linestyles= ['--', '-', '--'], colors= 'w')
    ax.set_title(f'C = {C}');
```

# Yeah, now it seems to be working...
